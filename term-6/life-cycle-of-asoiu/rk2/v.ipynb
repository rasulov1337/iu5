```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def task_V():
    sample1 = np.array([92, 105, 106, 106, 103, 92, 104, 106, 105, 100,
                         102, 114, 102, 122, 122, 99, 102, 114, 102, 99])  # Данные из Б
    sample2 = np.array([122.69, 112.79, 104.48, 106.06, 104.78, 113.19, 111.77, 111.38, 106.92,
                         108.31, 108.93, 95.51, 110.75, 97.21, 113.22, 107.28, 106.39, 96.61])  # Данные из В
    
    stat, p_value = stats.ks_2samp(sample1, sample2)
    print(f"Критерий Колмогорова-Смирнова: p-value={p_value}")
    if p_value > 0.05:
        print("Выборки не различимы")
    else:
        print("Выборки различаются")
    
    # График CDF
    sns.ecdfplot(sample1, label="Выборка Б")
    sns.ecdfplot(sample2, label="Выборка В")
    plt.legend()
    plt.title("График CDF")
    plt.show()

    # Полигон частот
    plt.figure(figsize=(8, 5))
    sns.histplot(sample1, bins=10, kde=False, stat="density", label="Выборка Б", color="blue", element="step")
    sns.histplot(sample2, bins=10, kde=False, stat="density", label="Выборка В", color="red", element="step")
    plt.xlabel("Значение")
    plt.ylabel("Плотность")
    plt.title("Полигон частот (эмпирическая плотность)")
    plt.legend()
    plt.show()
```

```python
task_V()
```
